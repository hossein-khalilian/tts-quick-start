```python
import re
import os
import requests
from IPython.display import Audio, display
```

```python
# GitHub API URL for the release by tag
url = "https://api.github.com/repos/k2-fsa/sherpa-onnx/releases/tags/tts-models"

# Send GET request
response = requests.get(url)

# Check if request was successful
tts_models = {}
if response.status_code == 200:
    data = response.json()
    assets = data.get('assets', [])
    
    print("Assets found:")
    for asset in assets:
        if re.search(r'-fa[-_]', asset["name"]):
            tts_models[asset["name"].split(".")[0]] = asset["browser_download_url"]
            print(f"- {asset['name']}")
            print(f"  Download URL: {asset['browser_download_url']}\n")
else:
    print(f"Failed to fetch release info: {response.status_code}")
```

```python
url = "http://host.docker.internal:5000/speech/generate-speech"
tts_model = "vits-piper-fa_IR-gyro-medium"
```

```python
text = "سلام! صبح شما هم بخیر و پر از طراوت. "
text += "امروز هوا آفتابیه ولی یکم گرمه"
```

```python
text
```

```python
payload = {
    "text": text,
    "model": tts_model,
    # "speed": 1.0
}
headers = {"Content-Type": "application/json"}

response = requests.post(url, json=payload, headers=headers)

# Step 2: Check and parse response
if response.status_code == 200:
    data = response.json()
    file_path = data.get("file_url").replace("dev", "jovyan")

    if file_path and os.path.exists(file_path):
        print(f"Playing audio: {file_path}")
        display(Audio(file_path))
    else:
        print("Audio file not found at path:", file_path)
else:
    print("Request failed:", response.status_code, response.text)
```
